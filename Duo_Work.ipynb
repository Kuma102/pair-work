# Задание 2.

```python
# загружаем библиотеки
from sklearn import metrics
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
```

```python
x=df[['Income', 'MntWines','MntMeatProducts','NumStorePurchases']]
x.head()
```

В переменную х были записаны столбцы Income, MntWines, MntMeatProducts,	NumStorePurchases

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
# Визуализируем данные
plt.scatter(x_scaled[:,0],x_scaled[:,2])
plt.xlabel('Income')
plt.ylabel('MntMeatProducts')
plt.show()
```

```python
# выведем количество возможных кластеров
from sklearn.cluster import KMeans
wcss = []
for k in range (1,11):
    Kmeans = KMeans(n_clusters=k, random_state=42)
    Kmeans.fit(x_scaled)
    wcss.append(Kmeans.inertia_)
plt.plot(range(1,11), wcss)
plt.xlabel('No of kmeans')
plt.ylabel('WCSS SCORE')
plt.show()
```

Исходя из графика делаем вывод, что лучше всего взять 3 кластера для нашего датасета

## KMeans

```python
# Обучение модели
Kmeans = KMeans(n_clusters=3,random_state=42)
kmeans_cluster = Kmeans.fit_predict(x_scaled)
df['kmeans_cluster'] = kmeans_cluster
```

```python
# Вывод случайных пяти строк
df.sample(5)
```

Модель обучилась и разбила датасет на два  кластера. Также в него добавился новый столбец 'kmeans_cluster', где записывается номер кластера

```python
# Визуализируем данные
plt.scatter(x_scaled[:,0],x_scaled[:,2], c=kmeans_cluster)
plt.xlabel('Income')
plt.ylabel('MntMeatProducts')
plt.show()
```

```python
sil=silhouette_score(x_scaled, kmeans_cluster)
dav=davies_bouldin_score(x_scaled, kmeans_cluster)
har=calinski_harabasz_score(x_scaled, kmeans_cluster)
print(f'Silhoutte score: {sil}')
print(f'Davis bouldin score: {dav}')
print(f'Calinski harabasz score: {har}')
```

Не очень хороший результат. Точки находятся на среднем расстоянии от центров кластеров. Сами кластеры не сильно плотные 

## Иерархическая кластеризация

```python
# Обучаем модель
from sklearn.cluster import AgglomerativeClustering

hc=AgglomerativeClustering(n_clusters=3, linkage='ward')
y_hc=hc.fit_predict(x_scaled)
```

```python
# Добавление столбца с кластерами
df['agglomerative_cluster']=y_hc
df.sample(5)
```

```python
# Визуализируем данные
plt.scatter(x_scaled[:,0],x_scaled[:,2], c=y_hc)
plt.xlabel('Income')
plt.ylabel('MntMeatProducts')
plt.show()
```

```python
sil=silhouette_score(x_scaled, y_hc)
dav=davies_bouldin_score(x_scaled, y_hc)
har=calinski_harabasz_score(x_scaled, y_hc)
print(f'Silhoutte score: {sil}')
print(f'Davis bouldin score: {dav}')
print(f'Calinski harabasz score: {har}')
```

Результат хуже чем в первом методе.

## DBSCAN

```python
# Обучение модели
from sklearn.cluster import DBSCAN

dbscan=DBSCAN(eps=0.3, min_samples=3)
dbscan_cluster=dbscan.fit_predict(x_scaled)
```

```python
# Добавление нового столбца с кластерами
df['dbscan_cluster']=dbscan_cluster
df.sample(5)
```

```python
# Визуализируем данные
plt.scatter(x_scaled[:,0],x_scaled[:,2], c=dbscan_cluster)
plt.xlabel('Income')
plt.ylabel('MntMeatProducts')
plt.show()
```

```python
sil=silhouette_score(x_scaled, dbscan_cluster)
dav=davies_bouldin_score(x_scaled, dbscan_cluster)
har=calinski_harabasz_score(x_scaled, dbscan_cluster)
print(f'Silhoutte score: {sil}')
print(f'Davis bouldin score: {dav}') 
print(f'Calinski harabasz score: {har}')
```

Самый худший результат

Среди всех методов, самый лучший результат показал метод KMeans. Дальше будем работать с ним.

# Задание 4

```python
# Загрузка библиотек
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix, classification_report, roc_curve, auc
from sklearn.preprocessing import label_binarize
```

```python
# объявдение целевой переменной
y=df['kmeans_cluster']
x=df[['Income', 'MntWines','MntMeatProducts','NumStorePurchases']]
```

```python
# объявление тестовых переменных
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.22,random_state=42)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

### Метод k-ближайщих соседей

```python
# Обучение модели
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train, y_train)
```

Модель обучана

```python
# Оценка качества обучения
y_pred_knn = knn.predict(x_test)
y_prob_knn = knn.predict_proba(x_test)
print('Метрики для KNN')
print(classification_report(y_test,y_pred_knn))
```

```python
knn_accuracy = accuracy_score(y_pred_knn, y_test)
knn_accuracy
```

```python
classes = [0, 1, 2]
y_test_bin = label_binarize(y_test, classes=classes)

plt.figure(figsize=(8, 6))

# Рисуем линию для каждого кластера по очереди
colors = ['purple', 'teal', 'gold'] 
for i, color in enumerate(colors):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_prob_knn[:, i])
    roc_auc = auc(fpr, tpr)
    
    plt.plot(fpr, tpr, color=color, lw=2, label=f'Кластер {i} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

Модель обучилась очень хорошо

## RandomForestClassifier

```python
# Обучение модели
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc.fit(x_train, y_train)
```

```python
# Оценка качества обучения модели
y_pred_rfc=rfc.predict(x_test)
y_prob_rfc=rfc.predict_proba(x_test)
print('Метрики для RandomForesClassifier')
print(classification_report(y_test,y_pred_rfc))
```

```python
rfc_accuracy = accuracy_score(y_pred_rfc, y_test)
rfc_accuracy
```

```python
classes = [0, 1, 2]
y_test_bin = label_binarize(y_test, classes=classes)

plt.figure(figsize=(8, 6))

# Рисуем линию для каждого кластера по очереди
for i, color in enumerate(colors):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_prob_rfc[:, i])
    roc_auc = auc(fpr, tpr)
    
    plt.plot(fpr, tpr, lw=2, label=f'Кластер {i} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

Результат слишком идеальный. Наша модель переучилась.

## SVC

```python
# Обучение моделей
from sklearn.svm import SVC
svc=SVC(probability=True)
svc.fit(x_train, y_train)
```

```python
classes = [0, 1, 2]
y_test_bin = label_binarize(y_test, classes=classes)

plt.figure(figsize=(8, 6))

# Рисуем линию для каждого кластера по очереди
for i, color in enumerate(colors):
    fpr, tpr, _ = roc_curve(y_test_bin[:, i], y_prob_svc[:, i])
    roc_auc = auc(fpr, tpr)
    
    plt.plot(fpr, tpr, lw=2, label=f'Кластер {i} (AUC = {roc_auc:.2f})')

plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.show()
```

Также очень хороший результат.

### Сравнение модели

```python
classifiers = ['KNeighborsClassifier', 'RandomForestClassifier', 'SVC']

log_cols = ['Classifier', 'Accuracy']
log = pd.DataFrame(columns=log_cols)

log["Classifier"] = classifiers
log["Accuracy"] = [knn_accuracy, rfc_accuracy, svc_accuracy]
log
```
